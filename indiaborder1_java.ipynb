```python
import java.awt.Color;
import java.awt.Dimension;
import java.awt.Graphics;
import java.awt.Graphics2D;
import java.awt.geom.Path2D;
import javax.swing.JFrame;
import javax.swing.JPanel;

public class IndiaMapBorder extends JPanel {

    // Simplified, approximate coordinates for India's border
    private static final double[][] INDIA_COORDS = {
        {68.7, 34.0}, {74.2, 35.0}, {76.8, 37.0}, {78.8, 38.0}, // Northern border
        {80.2, 34.0}, {82.0, 31.0}, {84.5, 27.0}, {88.0, 27.0}, // Northeastern border
        {88.7, 26.5}, {92.0, 28.0}, {95.0, 28.0}, {97.3, 27.0}, // Eastern Himalayas
        {94.0, 23.0}, {92.0, 21.0}, {88.5, 21.5}, {87.0, 21.0}, // Bangladesh and Myanmar border
        {86.0, 18.0}, {84.5, 16.0}, {80.0, 14.0}, {79.5, 11.0}, // Bay of Bengal coastline
        {77.5, 8.0}, {74.5, 8.0}, {72.0, 10.0}, {72.5, 12.0}, // Southern tip and western coast
        {69.5, 21.0}, {68.5, 23.0}, {68.2, 25.0}, {68.7, 34.0}  // Western border (Gujarat to Kashmir)
    };

    @Override
    protected void paintComponent(Graphics g) {
        super.paintComponent(g);
        Graphics2D g2d = (Graphics2D) g;

        int panelWidth = getWidth();
        int panelHeight = getHeight();

        // Calculate scaling and offset to fit the map in the panel
        double minLat = 8.0, maxLat = 38.0;
        double minLon = 68.0, maxLon = 98.0;
        double latRange = maxLat - minLat;
        double lonRange = maxLon - minLon;

        double scaleX = panelWidth / lonRange;
        double scaleY = panelHeight / latRange;
        double scale = Math.min(scaleX, scaleY);

        double offsetX = (panelWidth - lonRange * scale) / 2 - minLon * scale;
        double offsetY = (panelHeight - latRange * scale) / 2 - minLat * scale;

        // Use Path2D for smooth, connected lines
        Path2D path = new Path2D.Double();

        // Move to the first coordinate
        double startX = (INDIA_COORDS[0][0] * scale) + offsetX;
        double startY = panelHeight - ((INDIA_COORDS[0][1] * scale) + offsetY);
        path.moveTo(startX, startY);

        // Draw lines to all subsequent coordinates
        for (int i = 1; i < INDIA_COORDS.length; i++) {
            double nextX = (INDIA_COORDS[i][0] * scale) + offsetX;
            double nextY = panelHeight - ((INDIA_COORDS[i][1] * scale) + offsetY);
            path.lineTo(nextX, nextY);
        }

        g2d.setColor(Color.BLACK);
        g2d.draw(path);
    }

    @Override
    public Dimension getPreferredSize() {
        return new Dimension(800, 600);
    }

    public static void main(String[] args) {
        JFrame frame = new JFrame("India Map Border");
        frame.setDefaultCloseOperation(JFrame.EXIT_ON_CLOSE);
        frame.add(new IndiaMapBorder());
        frame.pack();
        frame.setLocationRelativeTo(null);
        frame.setVisible(true);
    }
}
```

```python
%%writefile IndiaMapBorder.java
import java.awt.*;
import java.awt.geom.Path2D;
import java.awt.image.BufferedImage;
import javax.imageio.ImageIO;
import java.io.File;

public class IndiaMapBorder {

    // Approximate coordinates for India's border
    private static final double[][] INDIA_COORDS = {
        {68.7, 34.0}, {74.2, 35.0}, {76.8, 37.0}, {78.8, 38.0},
        {80.2, 34.0}, {82.0, 31.0}, {84.5, 27.0}, {88.0, 27.0},
        {88.7, 26.5}, {92.0, 28.0}, {95.0, 28.0}, {97.3, 27.0},
        {94.0, 23.0}, {92.0, 21.0}, {88.5, 21.5}, {87.0, 21.0},
        {86.0, 18.0}, {84.5, 16.0}, {80.0, 14.0}, {79.5, 11.0},
        {77.5, 8.0}, {74.5, 8.0}, {72.0, 10.0}, {72.5, 12.0},
        {69.5, 21.0}, {68.5, 23.0}, {68.2, 25.0}, {68.7, 34.0}
    };

    public static void main(String[] args) throws Exception {

        int width = 800;
        int height = 600;

        BufferedImage image = new BufferedImage(width, height, BufferedImage.TYPE_INT_RGB);
        Graphics2D g2d = image.createGraphics();

        g2d.setColor(Color.WHITE);
        g2d.fillRect(0, 0, width, height);

        g2d.setColor(Color.BLACK);

        double minLat = 8.0, maxLat = 38.0;
        double minLon = 68.0, maxLon = 98.0;

        double latRange = maxLat - minLat;
        double lonRange = maxLon - minLon;

        double scaleX = width / lonRange;
        double scaleY = height / latRange;
        double scale = Math.min(scaleX, scaleY);

        double offsetX = (width - lonRange * scale) / 2 - minLon * scale;
        double offsetY = (height - latRange * scale) / 2 - minLat * scale;

        Path2D border = new Path2D.Double();

        double startX = (INDIA_COORDS[0][0] * scale) + offsetX;
        double startY = height - ((INDIA_COORDS[0][1] * scale) + offsetY);
        border.moveTo(startX, startY);

        for (int i = 1; i < INDIA_COORDS.length; i++) {

            double x = (INDIA_COORDS[i][0] * scale) + offsetX;
            double y = height - ((INDIA_COORDS[i][1] * scale) + offsetY);

            border.lineTo(x, y);
        }

        g2d.draw(border);
        g2d.dispose();

        ImageIO.write(image, "png", new File("india_map.png"));

        System.out.println("Map generated: india_map.png");
    }
}
```
